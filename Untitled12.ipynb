```python
!pip install pandas
import pandas as pd
!pip install numpy
import numpy as np
!pip install matplotlib
import matplotlib.pyplot as plt
!pip install sklearn
from sklearn.model_selection import train_test_split
from sklearn.linear_model import LinearRegression
# Load the CSV file into a DataFrame
df = pd.read_csv('/content/Final_ADME.csv')

# Display the first few rows of the DataFrame
print(df.head())
# Check for duplicates
duplicates = df.duplicated()
df_cleaned = df.drop_duplicates()
print(df_cleaned.head())





```

```python
#Data preparation
#Data separation as X and y
y= y = df['Ali Log S']
y
X = df.drop('Ali Log S', axis=1)
X

```

Data splitting

```python
from sklearn.model_selection import train_test_split


X_train, X_test, y_train, y_test = train_test_split(X, y, test_size=0.2, random_state=100)
```

```python
X_train
```

```python
X_test
```

Model Building

Training the model

Linear Regression

Applying the model to make a prediction

```python
import pandas as pd
from sklearn.linear_model import LinearRegression
from sklearn.impute import SimpleImputer

# Handle missing values in X_train
imputer = SimpleImputer(strategy='mean')  # Replace missing values
X_train_imputed = imputer.fit_transform(X_train.select_dtypes(include=['number'])) # Only impute numeric col
# Handle missing values in X_test
X_test_imputed = imputer.fit_transform(X_test.select_dtypes(include=['number'])) # Only impute numeric col

lr = LinearRegression()
lr.fit(X_train_imputed, y_train)  # Fit the model with imputed data
```

Applying the model to make a prediction

```python
y_lr_train_pred = lr.predict(X_train_imputed) # Use the imputed training data for prediction
y_lr_test_pred = lr.predict(X_test_imputed) # Use the imputed testing data
```

```python
y_lr_train_pred
```

```python
y_lr_test_pred
```

Evaluate model performance

```python
from sklearn.metrics import mean_squared_error, r2_score

lr_train_mse = mean_squared_error(y_train, y_lr_train_pred)
lr_train_r2 = r2_score(y_train, y_lr_train_pred)

lr_test_mse = mean_squared_error(y_test, y_lr_test_pred)
lr_test_r2 = r2_score(y_test, y_lr_test_pred)
```

```python
print('LR MSE (Train): ', lr_train_mse)
print('LR R2 (Train): ', lr_train_r2)
print('LR MSE (Test): ', lr_test_mse)
print('LR R2 (Test): ', lr_test_r2)
```

```python
lr_results = pd.DataFrame(['Linear regression', lr_train_mse, lr_train_r2, lr_test_mse, lr_test_r2]).transpose()
lr_results.columns = ['Method', 'Training MSE', 'Training R2', 'Test MSE', 'Test R2']
```
